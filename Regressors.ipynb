```python
from sklearn.dummy import DummyRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error, median_absolute_error, roc_auc_score, r2_score, explained_variance_score
from lightgbm import LGBMRegressor

modelnames = [
 'LGBMRegressor(n_estimators = 500)', 
 'KNeighborsRegressor()',
 'LinearRegression()', 
#'LogisticRegression()', 
 'SVR()',
 #'MLPRegressor(hidden_layer_sizes = (16, 16))',
 #'DecisionTreeRegressor()', 
 'GradientBoostingRegressor(n_estimators = 500)',
 'XGBRegressor(n_estimators = 500, max_depth=20)',
 'CatBoostRegressor(silent = True, n_estimators = 500)',
 'RandomForestRegressor(n_estimators = 500)',
]

metricnames = [
 'mean_absolute_error', 
# 'median_absolute_error',
 'r2_score',
# 'explained_variance_score'
]

columnnames = [
  'iter',  
  'assinaturaId',
 'assinatura', 
 'modelo', 
 'fold',
 'mean_absolute_error', 
 #'median_absolute_error',
 'r2_score',
 #'explained_variance_score'
]

metrics = pd.DataFrame(index = modelnames, columns = columnnames)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import numpy as np
resultados = []

for k in range(0,5):
    for i in range(0,len(assinaturas.columns)):
        print(assinaturas.columns[i])
        y = np.array(assinaturas[assinaturas.columns[i]])# Remove the labels from the features
        X = np.array(featurespd)

        X = X[~np.isnan(y)]
        y = y[~np.isnan(y)]

        for modelname in modelnames:
            #try:
                kf = KFold(n_splits=10,shuffle=True)
                kf.get_n_splits(X)
                for j, (train, test) in enumerate(kf.split(X)):
                  model = eval(modelname)
                  linha = []
                  pred_test = model.fit(X[train], y[train]).predict(X[test])
                  metrics.loc[modelname, 'assinaturaId'] = i
                  metrics.loc[modelname, 'assinatura'] = assinaturas.columns[i]
                  linha.append(k)
                  linha.append(i)
                  linha.append(assinaturas.columns[i])
                  linha.append(modelname)
                  linha.append(j)
                  for metricname in metricnames:
                    val = eval(f'{metricname}(y[test], pred_test)')
                    linha.append(val)
                    #print(val)
                    #metrics.loc[modelname, metricname] = eval(f'{metricname}(y_test, pred_test)')
                  resultados.append((linha))
            #except:
            #    print(modelname)
        rf = pd.DataFrame(resultados, columns=columnnames)
        rf.to_csv('results.csv',sep=";")  
    rf = pd.DataFrame(resultados, columns=columnnames)
    rf.to_csv('results.csv',sep=";")  
    
rf = pd.DataFrame(resultados, columns=columnnames)
```
